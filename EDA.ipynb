Esta clasificación será útil para determinar qué tipo de análisis estadístico y visualizaciones son apropiados para cada variable en el análisis exploratorio de datos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Markdown, display
import warnings
warnings.filterwarnings('ignore')

# Configuración para mejores visualizaciones
# plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

# Cargar los datos
car_data = pd.read_csv('datasets/vehicle_dataset/car_data.csv')
car_dekho = pd.read_csv('datasets/vehicle_dataset/CAR_DETAILS_FROM_CAR_DEKHO.csv')
car_v3 = pd.read_csv('datasets/vehicle_dataset/Car_details_v3.csv')
car_v4 = pd.read_csv('datasets/vehicle_dataset/car_details_v4.csv')

# Función para realizar análisis de variables categóricas
def analizar_categoricas(dataframe, nombre_df):
    print(f"\n## Análisis de Variables Categóricas - {nombre_df}")
    
    # Identificar variables categóricas
    cat_vars = dataframe.select_dtypes(include=['object']).columns
    
    for var in cat_vars:
        print(f"\n### Variable: {var}")
        
        # Análisis de frecuencias
        freq = dataframe[var].value_counts()
        freq_rel = dataframe[var].value_counts(normalize=True) * 100
        
        # Crear dataframe para mostrar frecuencias
        df_freq = pd.DataFrame({
            'Frecuencia Absoluta': freq,
            'Frecuencia Relativa (%)': freq_rel.round(2)
        })
        
        print("#### Análisis de Frecuencias")
        display(df_freq.head(10))
        
        # Análisis de cardinalidad
        cardinalidad = len(freq)
        print(f"#### Análisis de Cardinalidad")
        print(f"Cardinalidad: {cardinalidad} valores únicos")
        
        # Si la cardinalidad es muy alta, mostrar solo los valores más frecuentes
        if cardinalidad > 20:
            print(f"Mostrando solo los 10 valores más frecuentes de {cardinalidad} valores únicos.")

# Función para realizar análisis de variables numéricas
def analizar_numericas(dataframe, nombre_df):
    print(f"\n## Análisis de Variables Numéricas - {nombre_df}")
    
    # Identificar variables numéricas
    num_vars = dataframe.select_dtypes(include=['int64', 'float64']).columns
    
    # Calcular estadísticas descriptivas
    desc_stats = dataframe[num_vars].describe().T
    
    # Agregar desviación estándar si no está presente
    if 'std' not in desc_stats.columns:
        desc_stats['std'] = dataframe[num_vars].std()
    
    # Reordenar columnas para tener el formato solicitado
    desc_stats = desc_stats[['min', '50%', 'mean', 'std', 'max']]
    
    # Renombrar columnas
    desc_stats.columns = ['Mínimo', 'Mediana', 'Media', 'Desviación Típica', 'Máximo']
    
    # Redondear para mejor visualización
    desc_stats = desc_stats.round(2)
    
    print("### Estadísticas Descriptivas")
    display(desc_stats)
    
    # Análisis de distribución para variables numéricas seleccionadas
    print("\n### Distribución de Variables Numéricas Clave")
    
    # Seleccionar algunas variables numéricas para mostrar distribución
    if len(num_vars) > 0:
        key_vars = list(num_vars)[:min(3, len(num_vars))]
        
        print(f"Mostrando distribución para: {', '.join(key_vars)}")
        print("Para visualizar las distribuciones, ejecutar el código que genera histogramas")

# Aplicar análisis a cada conjunto de datos
print("# 1. Dataset: car_data.csv")
analizar_categoricas(car_data, "car_data")
analizar_numericas(car_data, "car_data")

print("\n\n# 2. Dataset: CAR_DETAILS_FROM_CAR_DEKHO.csv")
analizar_categoricas(car_dekho, "car_dekho")
analizar_numericas(car_dekho, "car_dekho")

print("\n\n# 3. Dataset: Car_details_v3.csv")
analizar_categoricas(car_v3, "car_v3")
analizar_numericas(car_v3, "car_v3")

print("\n\n# 4. Dataset: car_details_v4.csv")
analizar_categoricas(car_v4, "car_v4")
analizar_numericas(car_v4, "car_v4")
```

```python
# Visualizaciones para el EDA

# 1. Visualizaciones de distribuciones para variables numéricas
def visualizar_distribuciones(dataframe, nombre_df):
    print(f"\n## Visualizaciones de Distribuciones - {nombre_df}")
    
    # Seleccionar variables numéricas
    num_vars = dataframe.select_dtypes(include=['int64', 'float64']).columns
    
    if len(num_vars) > 0:
        # Crear subplots para histogramas
        fig, axes = plt.subplots(nrows=min(3, len(num_vars)), ncols=1, figsize=(12, 4*min(3, len(num_vars))))
        
        # Si solo hay una variable, axes no será un array
        if len(num_vars) == 1:
            axes = [axes]
        
        for i, var in enumerate(num_vars[:min(3, len(num_vars))]):
            # Histograma con curva de densidad KDE
            sns.histplot(dataframe[var], kde=True, ax=axes[i])
            axes[i].set_title(f'Distribución de {var}')
            axes[i].set_xlabel(var)
            axes[i].set_ylabel('Frecuencia')
            
        plt.tight_layout()
        plt.show()
        
        # Boxplots para variables numéricas
        fig, axes = plt.subplots(nrows=1, ncols=min(3, len(num_vars)), figsize=(15, 5))
        
        if len(num_vars[:min(3, len(num_vars))]) == 1:
            axes = [axes]
            
        for i, var in enumerate(num_vars[:min(3, len(num_vars))]):
            sns.boxplot(y=dataframe[var], ax=axes[i])
            axes[i].set_title(f'Boxplot de {var}')
            
        plt.tight_layout()
        plt.show()

# 2. Visualizaciones para variables categóricas
def visualizar_categoricas(dataframe, nombre_df):
    print(f"\n## Visualizaciones de Variables Categóricas - {nombre_df}")
    
    # Seleccionar variables categóricas
    cat_vars = dataframe.select_dtypes(include=['object']).columns
    
    for var in cat_vars:
        if dataframe[var].nunique() < 15:  # Solo mostrar variables con cardinalidad manejable
            plt.figure(figsize=(10, 6))
            # Gráfico de barras
            order = dataframe[var].value_counts().index
            ax = sns.countplot(x=var, data=dataframe, order=order)
            ax.set_title(f'Distribución de {var}')
            ax.set_xlabel(var)
            ax.set_ylabel('Frecuencia')
            
            # Rotar etiquetas si son muchas
            if dataframe[var].nunique() > 5:
                plt.xticks(rotation=45, ha='right')
                
            plt.tight_layout()
            plt.show()

# 3. Relación entre variables categóricas y precio
def visualizar_cat_vs_precio(dataframe, nombre_df, precio_col):
    print(f"\n## Visualizaciones de Variables Categóricas vs {precio_col} - {nombre_df}")
    
    # Seleccionar variables categóricas
    cat_vars = dataframe.select_dtypes(include=['object']).columns
    
    # Para cada variable categórica, mostrar boxplot vs precio
    for var in cat_vars:
        if dataframe[var].nunique() < 10:  # Solo mostrar variables con cardinalidad manejable
            plt.figure(figsize=(12, 6))
            
            order = dataframe.groupby(var)[precio_col].median().sort_values(ascending=False).index
            ax = sns.boxplot(x=var, y=precio_col, data=dataframe, order=order)
            ax.set_title(f'{var} vs {precio_col}')
            
            # Rotar etiquetas si son muchas
            if dataframe[var].nunique() > 5:
                plt.xticks(rotation=45, ha='right')
                
            plt.tight_layout()
            plt.show()

# 4. Matriz de correlación
def visualizar_correlacion(dataframe, nombre_df):
    print(f"\n## Matriz de Correlación - {nombre_df}")
    
    # Seleccionar variables numéricas
    num_vars = dataframe.select_dtypes(include=['int64', 'float64']).columns
    
    if len(num_vars) > 1:
        # Calcular correlación
        corr = dataframe[num_vars].corr()
        
        # Visualizar mapa de calor
        plt.figure(figsize=(10, 8))
        mask = np.triu(np.ones_like(corr, dtype=bool))
        cmap = sns.diverging_palette(230, 20, as_cmap=True)
        
        sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, vmin=-1, center=0,
                    square=True, linewidths=.5, annot=True, fmt=".2f", cbar_kws={"shrink": .8})
        
        plt.title(f'Matriz de Correlación - {nombre_df}')
        plt.tight_layout()
        plt.show()

# 5. Gráficos de dispersión para relaciones clave
def visualizar_dispersion(dataframe, nombre_df, precio_col, año_col, km_col):
    print(f"\n## Gráficos de Dispersión - {nombre_df}")
    
    cols_disponibles = []
    if precio_col in dataframe.columns:
        cols_disponibles.append(precio_col)
    if año_col in dataframe.columns:
        cols_disponibles.append(año_col)
    if km_col in dataframe.columns:
        cols_disponibles.append(km_col)
    
    if len(cols_disponibles) < 2:
        print("No hay suficientes variables clave disponibles para análisis de dispersión.")
        return
    
    # Crear gráficos de dispersión
    if año_col in dataframe.columns and precio_col in dataframe.columns:
        plt.figure(figsize=(10, 6))
        sns.regplot(x=año_col, y=precio_col, data=dataframe, scatter_kws={"alpha":0.4}, line_kws={"color":"red"})
        plt.title(f'Relación entre {año_col} y {precio_col}')
        plt.tight_layout()
        plt.show()
    
    if km_col in dataframe.columns and precio_col in dataframe.columns:
        plt.figure(figsize=(10, 6))
        sns.regplot(x=km_col, y=precio_col, data=dataframe, scatter_kws={"alpha":0.4}, line_kws={"color":"red"})
        plt.title(f'Relación entre {km_col} y {precio_col}')
        plt.tight_layout()
        plt.show()

# Aplicar visualizaciones para cada dataset
# Dataset 1: car_data
visualizar_distribuciones(car_data, "car_data")
visualizar_categoricas(car_data, "car_data")
visualizar_cat_vs_precio(car_data, "car_data", "Selling_Price")
visualizar_correlacion(car_data, "car_data")
visualizar_dispersion(car_data, "car_data", "Selling_Price", "Year", "Kms_Driven")

# Dataset 2: car_dekho
visualizar_distribuciones(car_dekho, "car_dekho")
visualizar_categoricas(car_dekho, "car_dekho")
visualizar_cat_vs_precio(car_dekho, "car_dekho", "selling_price")
visualizar_correlacion(car_dekho, "car_dekho")
visualizar_dispersion(car_dekho, "car_dekho", "selling_price", "year", "km_driven")

# Dataset 3: car_v3
visualizar_distribuciones(car_v3, "car_v3")
visualizar_categoricas(car_v3, "car_v3")
visualizar_cat_vs_precio(car_v3, "car_v3", "selling_price")
visualizar_correlacion(car_v3, "car_v3")
visualizar_dispersion(car_v3, "car_v3", "selling_price", "year", "km_driven")

# Dataset 4: car_v4
visualizar_distribuciones(car_v4, "car_v4")
visualizar_categoricas(car_v4, "car_v4")
visualizar_cat_vs_precio(car_v4, "car_v4", "Price")
visualizar_correlacion(car_v4, "car_v4")
visualizar_dispersion(car_v4, "car_v4", "Price", "Year", "Kilometer")
```
